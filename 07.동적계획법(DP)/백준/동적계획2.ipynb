# 1904

```python
# case1
N = int(input())

def F(N):
    if N == 1:
        return 1
    elif N == 2:
        return 2
    else:
        return F(N-2) + F(N-1)
    
print(F(N)%15746)
```

```python
# case2
N = int(input())

def F(N):
    if N <= 2: return N
    
    a = 1  # F[1]
    b = 2  # F[2]
    
    for _ in range(3, N+1):   # F(3)부터 시작
        n = a + b  # n = F(3)
        result = n
        a = b    # F[1]역할은 이제 F[2]가 하게 되고,
        b = n    # F[2]역할은 이제 F[3]가 하게 됨
        
    return result

print(F(N) % 15746)
```

```python
# case3
N = int(input())

F = [0 for i in range(1000001)]
F[1] = 1
F[2] = 2

for n in range(3, N+1):
    F[n] = F[n-2] + F[n-1]

print(F[N]%15746)
```

결과 값에만 나눠주는 게 아니라 반복문 안에서도 수시로 나머지 연산을 해 주어야 메모리 초과가 발생하지 않는다. (값이 int값을 초과하기 때문에 n = 1000000 일 경우 엄청나게 많은 메모리를 차지하게 된다.)

```python
# case4
N = int(input())

F = [0 for i in range(1000001)]
F[1] = 1
F[2] = 2

for i in range(3, N+1):
    F[i] = (F[i-2] + F[i-1]) % 15746

print(F[N])
```

```python
n = int(input())

F = [0] * 117
F[1] = 1
F[2] = 1
F[3] = 1

for i in range(3, n+1):
    F[i] = F[i-3] + F[i-1]

print(F[n])
```

# 1026

```python
n = int(input())
A = list(map(int, input().split()))
B = list(map(int, input().split()))

A.sort()
B.sort(reverse=True)

s = 0
for i in range(n):
    s = s + (A[i] * B[i])

print(s)
```
