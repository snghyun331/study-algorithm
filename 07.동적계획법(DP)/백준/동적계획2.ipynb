```python
# case1
N = int(input())

def F(N):
    if N == 1:
        return 1
    elif N == 2:
        return 2
    else:
        return F(N-2) + F(N-1)
    
print(F(N)%15746)
```

```python
# case2
N = int(input())

def F(N):
    if N <= 2: return N
    
    a = 1  # F[1]
    b = 2  # F[2]
    
    for _ in range(3, N+1):   # F(3) 
        n = a + b  # n = F(3)
        result = n
        a = b    # F[1]  F[2]  ,
        b = n    # F[2]  F[3]  
        
    return result

print(F(N) % 15746)
```

```python
# case3
N = int(input())

F = [0 for i in range(1000001)]
F[1] = 1
F[2] = 2

for n in range(3, N+1):
    F[n] = F[n-2] + F[n-1]

print(F[N]%15746)
```

# 1026

```python
n = int(input())
A = list(map(int, input().split()))
B = list(map(int, input().split()))

A.sort()
B.sort(reverse=True)

s = 0
for i in range(n):
    s = s + (A[i] * B[i])

print(s)
```
